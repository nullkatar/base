```python
import math as m

def u_an(t, r, teta):
    return r*m.cos(teta)*((5.0 - 4.0*t)**(-0.5))

if __name__ == '__main__':
    print ('Введите L')
    L = input()
    hr = 1.0/L*1.0
    print ('Введите M')
    M = input()
    hteta = m.pi/(M*2.0)
    print ('Введите N')
    N = input()
    tau = 1.0/N*1.0
        
    
    
    U = [[0] * (M+1) for i in range(L+1)]
    U_n = [[0] * (M+1) for i in range(L+1)]
    U_wave = [[0] * (M+1) for i in range(L+1)]
    U_nn = [[0] * (M+1) for i in range(L+1)]
    U_k = [[0] * (M+1) for i in range(L+1)]
    U_kk = [[0] * (M+1) for i in range(L+1)]
    
    
    a = [0 for i in range(0, max(L+1, M+1))]
    b = [0 for i in range(0, max(L+1, M+1))]
    c = [0 for i in range(0, max(L+1, M+1))]
    d = [0 for i in range(0, max(L+1, M+1))]
    A = [0 for i in range(0, max(L+1, M+1))]
    B = [0 for i in range(0, max(L+1, M+1))]
    
    
    for l in range(0, L+1):
        for i in range(0, M+1):
            U_k[l][i] = (hr*l)*m.cos(hteta*i)/m.sqrt(5.0)
            U_n[l][i] = (hr*l)*m.cos(hteta*i)/m.sqrt(5.0)
    
    for n in range(1, N+1):
        Max = 1
        while Max >= 0.0001:
            for i in range(0, M):
                a[0] = 0.
                b[0] = 1.
                c[0] = 0.
                d[0] = 0.
                a[L] = 0.
                b[L] = 1.
                c[L] = 0.
                d[L] = m.cos(hteta*i)/m.sqrt(5.0 - 4.0*tau*(n+1))
                for l in range(1, L):
                    a[l] = -(((l+0.5)*hr)**2)*tau*((U_k[l+1][i])**2 + (U_k[l][i])**2)/(2.0*(l**2)*(hr**4))
                    c[l] = -(((l-0.5)*hr)**2)*tau*((U_k[l][i])**2 + (U_k[l-1][i])**2)/(2.0*(l**2)*(hr**4))
                    b[l] = 1.0 - a[l] - c[l]
                    d[l] = U_n[l][i]
                    
                A[0] = -(a[0]/b[0])
                B[0] = d[0]/b[0]
                for l in range(1, L):
                    A[l] = -a[l]/(b[l] + A[l-1]*c[l])
                    B[l] = (d[l] - B[l-1]*c[l])/(b[l] + A[l-1]*c[l])
                    
                U_wave[0][i] = 0.0
                U_wave[L][i] = m.cos(hteta*i)/m.sqrt(5.0 - 4.0*tau*(n+1))
                for l in range (L-1, 0, -1):
                    U_wave[l][i] = A[l]*U_wave[l+1][i] + B[l]
            
            
            for l in range(1, L+1):
                a[0] = 0.
                b[0] = 1.
                c[0] = 0.
                d[0] = (hr*l)/m.sqrt(5.0 - 4.0*tau*(n+1))
                a[M] = 0.
                b[M] = 1.
                c[M] = 0.
                d[M] = 0.
                for i in range(1, M):
                    a[i] = -m.cos((i + 0.5)*hteta)*tau*(((U_k[l][i+1])**2) + ((U_k[l][i])**2))/(2.0*((hr*l)**2)*(hteta**2)*m.cos(i*hteta))
                    c[i] = -m.cos((i + 0.5)*hteta)*tau*(((U_k[l][i])**2) + ((U_k[l][i-1])**2))/(2.0*((hr*l)**2)*(hteta**2)*m.cos(i*hteta))
                    b[i] = 1.0 - a[i] - c[i]
                    d[i] = U_wave[l][i]
                    
                A[0] = -(a[0]/b[0])
                B[0] = d[0]/b[0]
                for i in range(1, M):
                    A[i] = -a[i]/(b[i] + A[i-1]*c[i])
                    B[i] = (d[i] - B[i-1]*c[i])/(b[i] + A[i-1]*c[i])
                
                U_kk[l][0] = (hr*l)/m.sqrt(5.0 - 4.0*tau*(n+1))
                U_kk[l][M] = 0.0
                for i in range (M-1, 0, -1):
                    U_kk[l][i] = A[i]*U_kk[l][i+1] + B[i]
            
            Max = 0
            for l in range(1, L):
                for i in range(1, M):
                    if Max < m.fabs((U_kk[l][i] - U_k[l][i])/U_kk[l][i]):
                        Max = m.fabs((U_kk[l][i] - U_k[l][i])/U_kk[l][i])
            
            for l in range(0, L+1):
                for i in range(0, M+1):
                    U_k[l][i] = U_kk[l][i]
            
        for l in range(0, L+1):
                for i in range(0, M+1):
                    U_n[l][i] = U_k[l][i]
                
    dmax = 0.0
    for l in range(0,11):
        for i in range(0,11):
            print 'r = ', l/10.0
            print 'teta = ', m.pi*i/20.0
            print 'U_an = ', u_an(1.0, l/10.0, m.pi*i/20.0)
            print 'U_cl = ', U_n[l*L/10][i*M/10]
            print 'd = ', m.fabs(u_an(1.0, l/10.0, m.pi*i/20.0) - U_n[l*L/10][i*M/10])
            print ' '
            if dmax < m.fabs(u_an(1.0, l/10.0, m.pi*i/20.0) - U_n[l*L/10][i*M/10]):
                dmax = m.fabs(u_an(1.0, l/10.0, m.pi*i/20.0) - U_n[l*L/10][i*M/10])       

        
    print 'dmax = ', dmax
```
